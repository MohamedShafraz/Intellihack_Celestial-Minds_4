```python
# Attempt to load the dataset with ISO-8859-1 encoding
data = pd.read_csv("/Users/sivagar/Downloads/Watches Bags Accessories.csv", encoding='ISO-8859-1')

# Display the first few rows of the dataset and its general information
data.head(), data.info()
```

```python
# Display unique values in the 'Rating in Stars' column to identify anomalies
data['Rating in Stars'].unique()
```

```python
# Display unique values in the 'Sold Count' column to identify format inconsistencies
data['Sold Count'].unique()
```

```python
# Function to convert 'Sold Count' to numerical value
def convert_sold_count(value):
    if pd.isna(value):
        return np.nan
    if 'K Sold' in value:
        return float(value.replace('K Sold', '')) * 1000
    if 'Sold' in value:
        return float(value.replace(' Sold', ''))
    return float(value)

# Apply conversion to 'Sold Count'
data['Sold Count'] = data['Sold Count'].apply(convert_sold_count)

# Fill missing 'Sold Count' with the median of the column
data['Sold Count'].fillna(data['Sold Count'].median(), inplace=True)

# Display the cleaned data and updated info
data.head(), data.info()
```

```python
import numpy as np

# Replace invalid ratings with NaN
invalid_ratings = ['5-Apr', '5-Mar', '5-May', '5-Jan', '0']
data['Rating in Stars'] = data['Rating in Stars'].replace(invalid_ratings, np.nan)

# Convert 'Rating in Stars' to numerical now that invalid entries are handled
data['Rating in Stars'] = data['Rating in Stars'].str.replace('/5', '').astype(float)

# Impute missing 'Rating in Stars' with the median of the column
data['Rating in Stars'].fillna(data['Rating in Stars'].median(), inplace=True)


```

```python
# Convert 'Current Price' and 'Original Price' to numerical values by removing currency symbols and converting to float
data['Current Price'] = data['Current Price'].str.extract('(\d+)').astype(float)
data['Original Price'] = data['Original Price'].str.extract('(\d+)').astype(float)

# Fill missing 'Original Price' values with the median
data['Original Price'].fillna(data['Original Price'].median(), inplace=True)

# Display the cleaned data and updated information
data.head(), data.info()
```
